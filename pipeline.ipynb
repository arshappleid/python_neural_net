```python
import math;
import numpy as np;
import pandas_datareader as dataRead;
import pandas as pd;
import seaborn as sns;
import matplotlib.pyplot as plt

## deep learning libraries
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
stockName= "TSLA"
startDate = "2021-03-01"  ## yyyy-mm-dd format
endDate = "2022-03-31"
df = pd.DataFrame(dataRead.DataReader(stockName, data_source='yahoo', start=startDate, end=endDate).round(2).reset_index())
df
```

```python
sns.set(rc = {'figure.figsize':(16,8)})
plot1 = sns.lineplot(x = df.Date , y = df.Close, data= "df" ).set( ylabel = "Close price in USD")
plot1
```
